### Create 500 random samples from normal distribution with mean 1 and  standard deviation 0.5

```python
import seaborn as sns
mu,stddev = 1,0.5
x=mu+stddev*np.random.randn(500)
print('The mean of 500 random sample with mean 1 and stand deviation 0.5:',x.mean().round(2))
```
